```python
x = 50

def func(x):
    print(f'X is {x}')

    # LOCAL REASSIGNMENT!
    x = 200
    print(f'I JUST LOCALLY CHANGE X TO {x}')
```

```python
func(x)
```

```python
print(x)
```

Si quisieramos definir alguna de las variables dentro de un metodo para volverla podemos hacer los siguiente:
Tambien debemos quitar el parametro de la funcion.

```python
x = 50

def func():
    global x
    print(f'X is {x}')

    # LOCAL REASSIGNMENT ON A GLOBAL VARIABLE!
    x = 'NEW VALUE'
    print(f'I JUST LOCALLY CHANGE X TO {x}')
```

```python
print(x)
```

```python
func()
```

```python
print(x)
```

Otra forma puedeser, sin usar la palabra clave global:

```python
x = 50

def func(x):
    print(f'X is {x}')

    # LOCAL REASSIGNMENT ON A GLOBAL VARIABLE!
    x = 'NEW VALUE'
    print(f'I JUST LOCALLY CHANGE X TO {x}')
    return x
```

```python
print(x)
```

```python
x = func(x)
```

```python
print(x)
```
